# Correlation between length of education and School enrollment rate.

### Throughout this project, I predict that there will be a strong correlation between the compulsory length of education and the school enrollment rate.


### The cell below imports pandas and reads the csv into a dataframe

```python
import pandas as pd

world_data = pd.read_csv("../data/databank_education_2018.csv")
education = world_data["SE.COM.DURS"]
enrollment = world_data["SE.PRM.NENR"]
```

### This next cell will take the Dataframes and prints the shape of the Dataframe and head.


```python
print(education.shape)
print(enrollment.shape)
world_data.head(10)
```

### This next cell will used column-based selection to crate a Dataframe with "education", "money", and "Country" and print a head of 10 rows.

```python
together = world_data[["Country","SE.COM.DURS","SE.PRM.NENR"]]
together.head(10)
```

### This next cell will used row-based filtering to create a Dataframe where both columns have values. 

```python
together_notna = together[education.notna() & enrollment.notna()]
print(together_notna.shape)
together_notna.head(10)
```

### This next cell will sort the values in descending order and display the 10 largest values.

```python
together_notna.sort_values(by="SE.PRM.NENR", ascending = False).head(10)
```

### This next cell produces a scatter plot of our data

```python
import numpy as np
x = np.polyfit(together_notna["SE.COM.DURS"], together_notna["SE.PRM.NENR"], 1)
y = np.poly1d(x)
together_notna.insert(1, "scatter", y(together_notna["SE.COM.DURS"]))
ax = together_notna.plot.scatter(x = "SE.COM.DURS", y = "SE.PRM.NENR")
together_notna.plot(x= "SE.COM.DURS", y="scatter",color= "Red", legend = False, ax=ax,  title = "Compulsory Education vs. School Enrollment Percentage")
ax.set_xlabel("Compulsory Education (years)")
ax.set_ylabel("School Enrollment percentage")
```

### This next cell prdocues a correlation between the two dataframes

```python
together_notna.corr()
```

### Overall due to the correlation (0.088692) being very close to 0, we can determine that there is a low to 0 correlation between the compulsory education and the school enrollment percentage of a country.
